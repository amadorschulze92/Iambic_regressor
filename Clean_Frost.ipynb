```python
import string
import numpy as np
import pandas as pd
```

```python
frost_sonnets = pd.read_csv('../../../Envs/ScrapeProj/sonnet_spider/frost.csv', delimiter= 'tabs')
frost_sonnets = pd.DataFrame(frost_sonnets)
frost_sonnets.columns = ['text']
frost_sonnets
```

```python
clean_frost = []
for x in frost_sonnets['text']:
    if '<dt>' in x or '<dd>' in x:
        clean_frost.append(x)
clean_frost
```

```python
clean_frost1 = []
author = []
for x in clean_frost:
    clean = str(x)
    clean = clean.replace('</p>','')
    clean = clean.replace('<dt>','')
    clean = clean.replace('-', '')
    clean = clean.replace('</dt>','')
    clean = clean.replace('<dd>','')
    clean = clean.replace('</dd>','')
    clean_frost1.append(clean)
    author.append('Frost')

frost_sonnets = pd.DataFrame({'text': clean_frost1,
                             'author': author})
```
