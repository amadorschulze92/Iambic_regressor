```python
import pandas as pd
```

```python
shakespeare = pd.read_csv('./assets/shake_sonnet_df.csv')
frost = pd.read_csv('./assets/frost_sonnet_df.csv')
keat = pd.read_csv('./assets/keat_sonnet_df.csv')
various = pd.read_csv('./assets/var_sonnet_df.csv')
```

```python
diff_df = [shakespeare, frost, keat, various]

sonnet_df = pd.concat(diff_df)

sonnet_df['sonnet'] = 1

sonnet_df.head(2)
```

```python
sonnet_df.to_csv('full_sonnet_df.csv')
```

```python
sonnet_df.info()
```

```python
austen = pd.read_csv('./assets/austen_text_df.csv')
doyle = pd.read_csv('./assets/doyle_text_df.csv')

diff_df = [austen, doyle]

text_df = pd.concat(diff_df)

text_df['sonnet'] = 0

text_df.head(2)
```

```python
text_df.info()
```

```python
text_df.to_csv('full_text_df.csv')
```

# Combine sonnets and text

```python
combine_df = [text_df, sonnet_df]

full_df = pd.concat(combine_df)
full_df.head()
```

```python
full_df.to_csv('final_df.csv')
```
